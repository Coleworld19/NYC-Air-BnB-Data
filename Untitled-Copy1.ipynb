```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NYC_AB = pd.read_csv('AB_NYC_2019.csv.zip')
NYC_AB.head()
```

```python
NYC_AB.drop(['id', 'latitude', 'longitude', 'calculated_host_listings_count'], axis = 1, inplace = True)
NYC_AB.head()
```

```python
NYC_AB.sort_values(by='price',axis=0, ascending = True, inplace=True, kind='quicksort', na_position='last')
NYC_AB
```

```python
# How many neighborhood groups are available and which shows up the most?
NYC_AB.groupby('neighbourhood_group').neighbourhood_group.count().sort_values(ascending=False)
```

```python
# How many neighborhood groups are available and which shows up the most?
NYC_AB.neighbourhood_group.value_counts().nlargest(1)
```

```python
# Are private rooms the most popular in manhattan?
NYC_AB.query('neighbourhood_group == "Manhattan"').room_type.value_counts().nlargest(5)
```

```python
NYC_AB.groupby('room_type')['room_type'].count()
```

```python
# Which hosts are the busiest based on their reviews?
NYC_AB.groupby('host_name').number_of_reviews.sum().nlargest(10)
```

```python
NYC_AB.host_name.value_counts()
```

```python
#Which neighorhood group has the highest average price?
neghb_group_highest_avg = NYC_AB.groupby('neighbourhood_group').price.mean().nlargest()
neghb_group_highest_avg
```

```python
# Which neighbor hood group has the highest total price?
neghb_group_highest_total = NYC_AB.groupby('neighbourhood_group').price.max().nlargest()
neghb_group_highest_total
```

```python
#Which top 5 hosts have the highest total price?
top5_hosts_high_total_price = NYC_AB.groupby('host_name')['price'].sum().nlargest(5)
top5_hosts_high_total_price
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
tophost_rev100_0avail = NYC_AB.query('number_of_reviews >= 100 and availability_365 == 0')
tophost_rev100_0avail
```

```python
# What host has the highest total of prices and where are they located?
host_high_price = NYC_AB.groupby(['host_name','neighbourhood']).price.sum().nlargest(1)
host_high_price
```

```python
sonder = NYC_AB.query('host_name == "Sonder (NYC)"')
sonder
```

```python
# When did Danielle from Queens last receive a review?
danielle_last_review = NYC_AB.query('host_name == "Danielle" and neighbourhood_group == "Queens"').sort_values('last_review', ascending=False)
danielle_last_review.head(1)
```
